## Points
- 토크나이저 = 토큰화 + ID 변환
- 실제 사용은 tokenizer(text) 한 줄이면 끝
- BERT는 서브워드 토크나이저(WordPiece) 사용

```python
# Load Tokenizer
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

# Tokenizer entire pipeline forward pass (tokenize + convert to id)
sentence = "Using a Transformer network is simple"

output = tokenizer(sentence)
print(output) # id, token type id, attention mask
```

Encoding

```python
# Only tokenization (Sub-word tokenization)
tokens = tokenizer.tokenize(sentence)
print(tokens)
```

```python
# Conversion to ids
ids = tokenizer.convert_tokens_to_ids(tokens)
print(ids)
```

Decoding

```python
decoded = tokenizer.decode(ids)
print(decoded)
```
